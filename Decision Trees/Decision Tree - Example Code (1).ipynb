```python
from sklearn.datasets import load_iris

iris = load_iris()
X, y = iris.data, iris.target
```

-- Viewing the data that we will work with.

```python
import pandas as pd
import numpy as np

pd.DataFrame(data= np.c_[iris['data'], iris['target']],columns= iris['feature_names'] + ['target'])
```

2. We specify the model using the following parameters: 


> Criterion: â€˜entropyâ€™ as opposed to the default â€˜Giniâ€™ that is used in the default decision trees.


If we ran the code as it is, it will create a decision tree that stops when each branch contains a 100% purity value, to stop the decision tree from making decision nodes, we can use the following parameters:


> max_depth: to stop the decision tree from exceeding a certain depth.


> min_impurity_decrease: to stop only when the impurity improvements are below a certain threshold


> min_samples_split: to stop only when the samples left are less than a certain number.


```python
from sklearn.tree import DecisionTreeClassifier

clf = DecisionTreeClassifier(criterion='entropy'
# , max_depth=2
# , min_impurity_decrease = 0.05
# , min_samples_split = 5
)
clf = clf.fit(X, y)
```

3. Visualize the tree

```python
import graphviz 
from sklearn import tree 

dot_data = tree.export_graphviz(clf, out_file=None, 
                     feature_names=iris.feature_names,  
                     class_names=iris.target_names,  
                     filled=True, rounded=True,  
                     special_characters=True)  
graph = graphviz.Source(dot_data)  
graph 
```
